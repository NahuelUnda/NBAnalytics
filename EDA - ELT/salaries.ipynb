# 1. Importación y carga de datos 

```python
import pandas as pd
```

```python
player = pd.read_csv('player.csv')
salaries = pd.read_csv('nba_2022-23_all_stats_with_salary.csv')
```

# 3. Filtro por los jugadores activos de player en salaries

```python
active_players = player[player['is_active'] == 1]
filtered_salaries = salaries[salaries['Player_name'].str.lower().isin(active_players['full_name'].str.lower())]

filtered_salaries
```

# 4. Agregado de a columna id de player en salaries

```python
filtered_salaries_id = filtered_salaries.merge(player[['id', 'full_name']], left_on='Player_name', right_on='full_name', how='left')
filtered_salaries_id.drop(columns=['full_name'], inplace=True)
filtered_salaries_id
```

# 5. Eliminación de Unnamed: 0 y ordenado de id

```python
filtered_salaries_id.drop(columns=['Unnamed: 0'], inplace=True)
filtered_salaries_id['id'] = filtered_salaries_id['id'].fillna(0).astype(int)
filtered_salaries_id.set_index('id', inplace=True)
filtered_salaries_id.reset_index(inplace=True)
filtered_salaries_id
```

# 6. Verificando nulos

```python
filtered_salaries_id.isnull().sum()[filtered_salaries_id.isnull().sum() > 0]
```

# 7. Llenado de nulos

## 7.1. Calculo de porcentaje para las columnas correspondientes

```python
filtered_salaries_id['FG_PCT'] = round((filtered_salaries_id['FG'] / filtered_salaries_id['FGA']).fillna(0), 2)
filtered_salaries_id['FG3_PCT'] = round((filtered_salaries_id['FG3'] / filtered_salaries_id['FG3A']).fillna(0), 2)
filtered_salaries_id['FG2_PCT'] = round((filtered_salaries_id['FG2'] / filtered_salaries_id['FG2A']).fillna(0), 2)
filtered_salaries_id['eFG_PCT'] = round(((filtered_salaries_id['FG'] + 0.5 * filtered_salaries_id['FG3']) / filtered_salaries_id['FGA']).fillna(0), 2)
filtered_salaries_id['FT_PCT'] = round((filtered_salaries_id['FT'] / filtered_salaries_id['FTA']).fillna(0), 2)
filtered_salaries_id['FT_PCT'] = round((filtered_salaries_id['FT'] / filtered_salaries_id['FTA']).fillna(0), 2)
filtered_salaries_id['TS_PCT'] = round((filtered_salaries_id['PTS'] / (2 * (filtered_salaries_id['FGA'] + 0.44 * filtered_salaries_id['FTA']))).fillna(0), 2)

filtered_salaries_id
```

```python
filtered_salaries_id.isnull().sum()[filtered_salaries_id.isnull().sum() > 0]
```

## 7.2. Calculos correspondientes para columnas faltantes

```python
filtered_salaries_id['FG3Ar'] = round((filtered_salaries_id['FG3A'] / filtered_salaries_id['FGA']).fillna(0), 2)

filtered_salaries_id
```

```python
filtered_salaries_id['FTr'] = round((filtered_salaries_id['FTA'] / filtered_salaries_id['FGA']).fillna(0), 2)

filtered_salaries_id
```

```python
filtered_salaries_id.isnull().sum()[filtered_salaries_id.isnull().sum() > 0]
```

# 8. Carga de variables de entorno

```python
from dotenv import load_dotenv
import os

# Carga variables desde el archivo .env
load_dotenv()

# Accede a las variables
SERVER = os.getenv("SERVER")
DATABASE = os.getenv("DATABASE")
USER = os.getenv("USER")
PASSWORD = os.getenv("PASSWORD")
DRIVER = os.getenv("DRIVER")
```

# 9. Conexion a la base de datos

```python
from sqlalchemy.exc import SQLAlchemyError
from sqlalchemy import create_engine

# Crear la cadena de conexión
connection_string = f"mssql+pyodbc://{USER}:{PASSWORD}@{SERVER}/{DATABASE}?driver={DRIVER}"
engine = create_engine(connection_string)

# Verificar la conexión
try:
    with engine.connect() as connection:
        print("Conexión a la base de datos exitosa.")
        print(f"Servidor: {SERVER}")
except Exception as e:
    print(f"Error al conectar a la base de datos: {e}")
```

# 10. Inserción tabla Salaries

```python
try:
	filtered_salaries_id.to_sql("Salaries", con=engine, if_exists="append", index=False)
	print("Datos insertados en la tabla 'Salaries'")
except SQLAlchemyError as e:
	print(f"A ocurrido un error: {e}")
```
